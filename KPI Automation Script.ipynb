```python
# Import Modules
from openpyxl import load_workbook
from docx import Document
from pathlib import Path
import os

# Automatically get retrieve Excel and Word files from the same folder as this Automation Script
BASE_DIR = Path(os.getcwd())

# -----------------------------
# FILE PATHS (Edit as needed)
# -----------------------------
xlsx_path = BASE_DIR / "KPI_LINKED_BRIEFING_NOTE_FEED_TEMPLATE.xlsx"
docx_path = BASE_DIR / "Briefing Note - KC Applications and Enrolments KPI Update_LINKED_TEMPLATE.docx"
out_path  = BASE_DIR / "Briefing Note - KC Applications and Enrolments KPI Update as of [ENTER EXTACTION DATE HERE].docx"

# -----------------------------
# MAPPING: Word placeholder -> (Excel sheet, Excel cell)
# Users edit THIS section only.
# -----------------------------
MAPPING = {
    # Header and Subject Mapping Variables
    "{{B4}}": ("Briefing Note Feed", "B4"),
    "{{B5}}": ("Briefing Note Feed", "B5"),
    "{{B6}}": ("Briefing Note Feed", "B6"),
    "{{B7}}": ("Briefing Note Feed", "B7"),
    "{{B8}}": ("Briefing Note Feed", "B8"),
    "{{B9}}": ("Briefing Note Feed", "B9"),
    "{{B10}}": ("Briefing Note Feed", "B10"),
    
    # Applications Briefing Note Mapping Variables
    "{{D3}}": ("Briefing Note Feed", "D3"),
    "{{E4}}": ("Briefing Note Feed", "E4"),
    "{{E5}}": ("Briefing Note Feed", "E5"),
    "{{E6}}": ("Briefing Note Feed", "E6"),
    "{{E7}}": ("Briefing Note Feed", "E7"),

    "{{F4}}": ("Briefing Note Feed", "F4"),
    "{{F5}}": ("Briefing Note Feed", "F5"),
    "{{F6}}": ("Briefing Note Feed", "F6"),
    "{{F7}}": ("Briefing Note Feed", "F7"),

    "{{G5}}": ("Briefing Note Feed", "G5"),
    "{{G6}}": ("Briefing Note Feed", "G6"),
    "{{G7}}": ("Briefing Note Feed", "G7"),

    "{{H4}}": ("Briefing Note Feed", "H4"),
    "{{H5}}": ("Briefing Note Feed", "H5"),
    "{{H6}}": ("Briefing Note Feed", "H6"),
    "{{H7}}": ("Briefing Note Feed", "H7"),

    "{{I4}}": ("Briefing Note Feed", "I4"),
    "{{I5}}": ("Briefing Note Feed", "I5"),
    "{{I6}}": ("Briefing Note Feed", "I6"),
    "{{I7}}": ("Briefing Note Feed", "I7"),

    "{{J5}}": ("Briefing Note Feed", "J5"),
    "{{J6}}": ("Briefing Note Feed", "J6"),
    "{{J7}}": ("Briefing Note Feed", "J7"),

    "{{D9}}": ("Briefing Note Feed", "D9"),
    "{{D10}}": ("Briefing Note Feed", "D10"),
    "{{D11}}": ("Briefing Note Feed", "D11"),
    "{{D12}}": ("Briefing Note Feed", "D12"),

    "{{D14}}": ("Briefing Note Feed", "D14"),
    "{{D15}}": ("Briefing Note Feed", "D15"),
    "{{D16}}": ("Briefing Note Feed", "D16"),
    "{{D17}}": ("Briefing Note Feed", "D17"),

    "{{D26}}": ("Briefing Note Feed", "D26"),
    "{{E27}}": ("Briefing Note Feed", "E27"),
    "{{E28}}": ("Briefing Note Feed", "E28"),
    "{{E29}}": ("Briefing Note Feed", "E29"),
    "{{E30}}": ("Briefing Note Feed", "E30"),

    "{{F27}}": ("Briefing Note Feed", "F27"),
    "{{F28}}": ("Briefing Note Feed", "F28"),
    "{{F29}}": ("Briefing Note Feed", "F29"),
    "{{F30}}": ("Briefing Note Feed", "F30"),

    "{{G28}}": ("Briefing Note Feed", "G28"),
    "{{G29}}": ("Briefing Note Feed", "G29"),
    "{{G30}}": ("Briefing Note Feed", "G30"),

    "{{H27}}": ("Briefing Note Feed", "H27"),
    "{{H28}}": ("Briefing Note Feed", "H28"),
    "{{H29}}": ("Briefing Note Feed", "H29"),
    "{{H30}}": ("Briefing Note Feed", "H30"),

    "{{I27}}": ("Briefing Note Feed", "I27"),
    "{{I28}}": ("Briefing Note Feed", "I28"),
    "{{I29}}": ("Briefing Note Feed", "I29"),
    "{{I30}}": ("Briefing Note Feed", "I30"),

    "{{J28}}": ("Briefing Note Feed", "J28"),
    "{{J29}}": ("Briefing Note Feed", "J29"),
    "{{J30}}": ("Briefing Note Feed", "J30"),

    "{{D32}}": ("Briefing Note Feed", "D32"),
    "{{D33}}": ("Briefing Note Feed", "D33"),
    "{{D34}}": ("Briefing Note Feed", "D34"),
    "{{D35}}": ("Briefing Note Feed", "D35"),

    "{{D37}}": ("Briefing Note Feed", "D37"),
    "{{D38}}": ("Briefing Note Feed", "D38"),
    "{{D39}}": ("Briefing Note Feed", "D39"),
    "{{D40}}": ("Briefing Note Feed", "D40"),

    # Enrolment Briefing Note Mapping Variables
    "{{D48}}": ("Briefing Note Feed", "D48"),
    "{{D49}}": ("Briefing Note Feed", "D49"),
    "{{F50}}": ("Briefing Note Feed", "F50"),
    "{{H50}}": ("Briefing Note Feed", "H50"),
    "{{F53}}": ("Briefing Note Feed", "F53"),
    "{{F54}}": ("Briefing Note Feed", "F54"),
    "{{F55}}": ("Briefing Note Feed", "F55"),
    "{{F56}}": ("Briefing Note Feed", "F56"),
    "{{F57}}": ("Briefing Note Feed", "F57"),
    "{{F58}}": ("Briefing Note Feed", "F58"),
    "{{F59}}": ("Briefing Note Feed", "F59"),
    "{{F60}}": ("Briefing Note Feed", "F60"),
    "{{F61}}": ("Briefing Note Feed", "F61"),
    "{{F62}}": ("Briefing Note Feed", "F62"),
    "{{G53}}": ("Briefing Note Feed", "G53"),
    "{{G54}}": ("Briefing Note Feed", "G54"),
    "{{G55}}": ("Briefing Note Feed", "G55"),
    "{{G56}}": ("Briefing Note Feed", "G56"),
    "{{G57}}": ("Briefing Note Feed", "G57"),
    "{{G58}}": ("Briefing Note Feed", "G58"),
    "{{G59}}": ("Briefing Note Feed", "G59"),
    "{{G60}}": ("Briefing Note Feed", "G60"),
    "{{G61}}": ("Briefing Note Feed", "G61"),
    "{{G62}}": ("Briefing Note Feed", "G62"),
    "{{H53}}": ("Briefing Note Feed", "H53"),
    "{{H54}}": ("Briefing Note Feed", "H54"),
    "{{H55}}": ("Briefing Note Feed", "H55"),
    "{{H56}}": ("Briefing Note Feed", "H56"),
    "{{H57}}": ("Briefing Note Feed", "H57"),
    "{{H58}}": ("Briefing Note Feed", "H58"),
    "{{H59}}": ("Briefing Note Feed", "H59"),
    "{{H60}}": ("Briefing Note Feed", "H60"),
    "{{H61}}": ("Briefing Note Feed", "H61"),
    "{{H62}}": ("Briefing Note Feed", "H62"),
    "{{I53}}": ("Briefing Note Feed", "I53"),
    "{{I54}}": ("Briefing Note Feed", "I54"),
    "{{I55}}": ("Briefing Note Feed", "I55"),
    "{{I56}}": ("Briefing Note Feed", "I56"),
    "{{I57}}": ("Briefing Note Feed", "I57"),
    "{{I58}}": ("Briefing Note Feed", "I58"),
    "{{I59}}": ("Briefing Note Feed", "I59"),
    "{{I60}}": ("Briefing Note Feed", "I60"),
    "{{I61}}": ("Briefing Note Feed", "I61"),
    "{{I62}}": ("Briefing Note Feed", "I62"),
    "{{J53}}": ("Briefing Note Feed", "J53"),
    "{{J54}}": ("Briefing Note Feed", "J54"),
    "{{J55}}": ("Briefing Note Feed", "J55"),
    "{{J56}}": ("Briefing Note Feed", "J56"),
    "{{J57}}": ("Briefing Note Feed", "J57"),
    "{{J58}}": ("Briefing Note Feed", "J58"),
    "{{J59}}": ("Briefing Note Feed", "J59"),
    "{{J60}}": ("Briefing Note Feed", "J60"),
    "{{J61}}": ("Briefing Note Feed", "J61"),
    "{{J62}}": ("Briefing Note Feed", "J62"),
    "{{K53}}": ("Briefing Note Feed", "K53"),
    "{{K54}}": ("Briefing Note Feed", "K54"),
    "{{K55}}": ("Briefing Note Feed", "K55"),
    "{{K56}}": ("Briefing Note Feed", "K56"),
    "{{K57}}": ("Briefing Note Feed", "K57"),
    "{{K58}}": ("Briefing Note Feed", "K58"),
    "{{K59}}": ("Briefing Note Feed", "K59"),
    "{{K60}}": ("Briefing Note Feed", "K60"),
    "{{K61}}": ("Briefing Note Feed", "K61"),
    "{{K62}}": ("Briefing Note Feed", "K62"),

    "{{D64}}": ("Briefing Note Feed", "D64"),
    "{{D65}}": ("Briefing Note Feed", "D65"),
    "{{D66}}": ("Briefing Note Feed", "D66"),
    "{{D67}}": ("Briefing Note Feed", "D67"),
    "{{D68}}": ("Briefing Note Feed", "D68"),
    "{{D69}}": ("Briefing Note Feed", "D69"),
    "{{D71}}": ("Briefing Note Feed", "D71"),
    "{{D72}}": ("Briefing Note Feed", "D72"),
    "{{D73}}": ("Briefing Note Feed", "D73"),
    
    "{{D75}}": ("Briefing Note Feed", "D75"),
    "{{F79}}": ("Briefing Note Feed", "F79"),
    "{{F80}}": ("Briefing Note Feed", "F80"),
    "{{F81}}": ("Briefing Note Feed", "F81"),
    "{{F82}}": ("Briefing Note Feed", "F82"),
    "{{G79}}": ("Briefing Note Feed", "G79"),
    "{{G80}}": ("Briefing Note Feed", "G80"),
    "{{G81}}": ("Briefing Note Feed", "G81"),
    "{{G82}}": ("Briefing Note Feed", "G82"),
    "{{H79}}": ("Briefing Note Feed", "H79"),
    "{{H80}}": ("Briefing Note Feed", "H80"),
    "{{H81}}": ("Briefing Note Feed", "H81"),
    "{{H82}}": ("Briefing Note Feed", "H82"),

    "{{L79}}": ("Briefing Note Feed", "L79"),
    "{{L80}}": ("Briefing Note Feed", "L80"),
    "{{L81}}": ("Briefing Note Feed", "L81"),
    "{{L82}}": ("Briefing Note Feed", "L82"),
    "{{M79}}": ("Briefing Note Feed", "M79"),
    "{{M80}}": ("Briefing Note Feed", "M80"),
    "{{M81}}": ("Briefing Note Feed", "M81"),
    "{{M82}}": ("Briefing Note Feed", "M82"),
    "{{N79}}": ("Briefing Note Feed", "N79"),
    "{{N80}}": ("Briefing Note Feed", "N80"),
    "{{N81}}": ("Briefing Note Feed", "N81"),
    "{{N82}}": ("Briefing Note Feed", "N82"),

    "{{D87}}": ("Briefing Note Feed", "D87"),
    "{{D88}}": ("Briefing Note Feed", "D88"),
    "{{D89}}": ("Briefing Note Feed", "D89"),

    "{{J79}}": ("Briefing Note Feed", "J79"),
    "{{J80}}": ("Briefing Note Feed", "J80"),
    "{{J81}}": ("Briefing Note Feed", "J81"),
    "{{J82}}": ("Briefing Note Feed", "J82"),

    "{{I79}}": ("Briefing Note Feed", "I79"),
    "{{I80}}": ("Briefing Note Feed", "I80"),
    "{{I81}}": ("Briefing Note Feed", "I81"),
    "{{I82}}": ("Briefing Note Feed", "I82"),

    "{{K79}}": ("Briefing Note Feed", "K79"),
    "{{K80}}": ("Briefing Note Feed", "K80"),
    "{{K81}}": ("Briefing Note Feed", "K81"),
    "{{K82}}": ("Briefing Note Feed", "K82"),

    "{{E83}}": ("Briefing Note Feed", "E83"),

    "{{D90}}": ("Briefing Note Feed", "D90"),
    "{{D91}}": ("Briefing Note Feed", "D91"),
    "{{D92}}": ("Briefing Note Feed", "D92"),

    "{{D94}}": ("Briefing Note Feed", "D94"),
    "{{D95}}": ("Briefing Note Feed", "D95"),
    "{{D96}}": ("Briefing Note Feed", "D96"),
    "{{D97}}": ("Briefing Note Feed", "D97"),
    "{{D98}}": ("Briefing Note Feed", "D98"),
    "{{D99}}": ("Briefing Note Feed", "D99"),

    "{{D102}}": ("Briefing Note Feed", "D102"),
    "{{D103}}": ("Briefing Note Feed", "D103"),

    "{{D106}}": ("Briefing Note Feed", "D106"),
    "{{D107}}": ("Briefing Note Feed", "D107"),

}

# -----------------------------
# Below this line is the mapping script, no need for user modifications beyond this point
# -----------------------------

# --- load Excel (data_only=True reads last saved values for formulas) ---
wb = load_workbook(xlsx_path, data_only=True)

# Build: placeholder -> string value
values = {}
for placeholder, (sheet_name, cell_addr) in MAPPING.items():
    v = wb[sheet_name][cell_addr].value
    values[placeholder] = "" if v is None else str(v)

doc = Document(docx_path)

def replace_all_placeholders_in_paragraph(paragraph, placeholder_to_value):
    """
    Replace any placeholders in this paragraph, even if split across runs,
    while preserving formatting (keeps formatting of the first run).
    """
    runs = paragraph.runs
    if not runs:
        return

    full = "".join(r.text for r in runs)

    # Quick check: skip if none of the placeholders are present
    if not any(ph in full for ph in placeholder_to_value):
        return

    new_full = full
    for ph, val in placeholder_to_value.items():
        new_full = new_full.replace(ph, val)

    if new_full == full:
        return

    runs[0].text = new_full
    for r in runs[1:]:
        r.text = ""

def process_paragraphs(paragraphs, placeholder_to_value):
    for p in paragraphs:
        replace_all_placeholders_in_paragraph(p, placeholder_to_value)

def process_tables(tables, placeholder_to_value):
    for table in tables:
        for row in table.rows:
            for cell in row.cells:
                process_paragraphs(cell.paragraphs, placeholder_to_value)

# 1) document body
process_paragraphs(doc.paragraphs, values)
process_tables(doc.tables, values)

# 2) headers & footers (all sections)
for section in doc.sections:
    process_paragraphs(section.header.paragraphs, values)
    process_tables(section.header.tables, values)

    process_paragraphs(section.footer.paragraphs, values)
    process_tables(section.footer.tables, values)

doc.save(out_path)
print(f"Saved: {out_path}")
```
